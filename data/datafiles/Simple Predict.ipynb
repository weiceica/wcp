```python
#normalize data for the specific train group
ArgentinaH_df = df2[df2['Home Team Name'] == "Argentina"] 
ArgentinaA_df = df2[df2['Away Team Name'] == 'Argentina']
Argentina_df = ArgentinaH_df.append(ArgentinaA_df, ignore_index=True)
BrazilH_df = df2[df2['Home Team Name'] == "Brazil"] 
BrazilA_df = df2[df2['Away Team Name'] == 'Brazil']
Brazil_df = BrazilH_df.append(BrazilA_df, ignore_index=True)
```

```python
Argentina_df
Brazil_df
```

```python
BrazilH_parsed = BrazilH_df[["Home Team Goals", "Away Team Goals"]]
```

```python
BrazilH_parsed
```

```python
train_df = BrazilH_parsed.sample(frac=0.75, random_state=4)
#drops training data from original dataframe, rest of 25% is validation/test data
val_df = BrazilH_parsed.drop(train_df.index)
```

```python
X_train = train_df.drop('Home Team Goals',axis=1)
X_val = val_df.drop('Home Team Goals',axis=1)

# target
y_train = train_df['Away Team Goals']
y_val = val_df['Away Team Goals']
 
# format the features/inputs as an argument for model by shaping it accordingly
input_shape = [X_train.shape[1]]
input_shape
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(units=64, activation='relu', input_shape=input_shape),
    tf.keras.layers.Dense(units=64, activation='relu'),
    tf.keras.layers.Dense(units=1)
])
model.summary()
```

```python
model.compile(optimizer='adam', loss='mae')
```

```python
losses = model.fit(X_train, y_train, validation_data=(X_val, y_val), batch_size=256, epochs=9)
```
